```python
import numpy as np
from scipy.interpolate import BSpline
import matplotlib.pyplot as plt

rmax = 400
dr = 0.01
r = np.arange(0,rmax+dr,dr)

N_knots = 50

order = 3

        
knots_start_end = np.repeat([0, rmax],order)  # Repeats -end and end, order times
knots_middle = np.linspace(0, rmax - 1, N_knots - 2 * order)  # Creates evenly spaced knots between the start and end
knots = np.concatenate([knots_start_end[:order], knots_middle, knots_start_end[order:]])  # Concatenates the start, middle, and end knots

basis_funcs = [BSpline(knots, (i == np.arange(len(knots) - order - 1 )).astype(float), order) for i in range(len(knots) - order - 1 )[1:-(order+1)]] 
n_basis = len(basis_funcs)

    
basis_array = np.empty((len(r),n_basis))

for i in range(n_basis):
        basis_array[:,i] = basis_funcs[i](r)
        
        
#for i in range(n_basis):
    #plt.plot(r,basis_array[:,i])
#plt.savefig("b_splines.png")
```

```python
from scipy.integrate import quad
from scipy.interpolate import BSpline
from scipy.integrate import simps
import time

bspline1 = basis_funcs[9]
bspline2 = basis_funcs[9]

# Calculate the common support
start = time.time()
common_support_min = max(bspline1.t[bspline1.k], bspline2.t[bspline2.k])
common_support_max = min(bspline1.t[-bspline1.k-1], bspline2.t[-bspline2.k-1])

# Define the integrand as a function that takes a single argument.
def integrand(r):
    return bspline1(r) * bspline2(r)

# Now perform the integration over the common support
integral, error = quad(integrand, common_support_min, common_support_max)
end = time.time()
print("Integral:", integral,"Time",end-start)

t1 = basis_funcs[9](r)
t2 = basis_funcs[9](r)
start = time.time()
I = simps(basis_funcs[9](r)*basis_funcs[9](r),r)
end = time.time()
print("Integral:", I,"Time",end-start)
```
