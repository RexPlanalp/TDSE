```python
from scipy.sparse import random as sparse_random
import numpy as np

# Generating two dense scipy sparse matrices
# Using a high density value to ensure the matrices are very dense
density = 0.50  # 90% of the elements will be non-zero

# Matrix dimensions
rows, cols = 100, 100

# Creating two dense matrices
matrix1 = sparse_random(rows, cols, density=density, format='csr', dtype=np.float64)
matrix2 = sparse_random(rows, cols, density=density, format='csr', dtype=np.float64)
import time
start = time.time()
matrix1 + matrix2
end = time.time()
print(end-start)
```
