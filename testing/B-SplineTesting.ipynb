```python
dr = 0.01
rmax = 1000
r = np.arange(0,rmax+dr,dr)

N_knots = 100
order = 12


linear = np.linspace(0,rmax/3,int(N_knots/2))


a = 10
x = np.linspace(np.sqrt((linear[-1])*a),np.sqrt(rmax*a),int(N_knots/2))
quadratic = x**2 /a
knots = np.append(linear,quadratic)

for _ in range(order+1):
    knots = np.insert(knots,0,0)
    knots = np.append(knots,rmax)

N_knots = len(knots)



n_basis = len(knots) - order - 1

basis_funcs = [BSpline(knots, (i == np.arange(n_basis)), order) for i in range(n_basis)[int(order/5):-int(order/5)-1]]


    
n_basis = len(basis_funcs)

print(n_basis)

for i in range(n_basis):
    plt.plot(r,basis_funcs[i](r))
    #plt.xlim([0,10])
    
```
