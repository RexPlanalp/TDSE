```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import sph_harm
```

```python
root = "//data/becker/dopl4670/TDSE_Jobs/Project/10_400_4.16E14/TDSE_files/"
file = "A_slice.npy"

A = np.load(root+file)

phi = np.arange(0, 2*np.pi+0.01, 0.01)
theta = np.pi/2

plt.plot(phi,A)

print(np.max(A))

print(phi[np.argmin(A)])


real_sph_24 = (sph_harm(8, 8, phi, theta)*np.exp(-1j*8*phi))[0]
real_sph_25 = (sph_harm(9, 9, phi, theta)*np.exp(-1j*9*phi))[0]



alpha = np.abs((real_sph_24**2 + real_sph_25**2) * np.max(A))/(2*real_sph_24*real_sph_25)
alpha = np.abs(alpha)


print(np.abs(alpha-np.max(A)))



```
