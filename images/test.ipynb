```python
import numpy as np
import matplotlib.pyplot as plt

PAD = np.load("PAD.npy")


E_vals = PAD[0,:]
theta_vals = PAD[1,:]
phi_vals = PAD[2,:]
pad_vals = PAD[3,:]
```

```python
test = pad_vals * np.sin(theta_vals)
```

```python
energy_dict = {}
for i,value in enumerate(test):
    if E_vals[i] in energy_dict:
        energy_dict[E_vals[i]] += value
    else:
        energy_dict[E_vals[i]] = 0
```

```python
energies = np.array(sorted(energy_dict.keys()))
probabilities = np.array([energy_dict[energy] for energy in energies])
```

```python
plt.semilogy(energies,probabilities*0.01)
```
